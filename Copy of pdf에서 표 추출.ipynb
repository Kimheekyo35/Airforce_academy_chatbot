```python
import pdfplumber
import json
import pandas as pd


def extract_table(pdf_path):
  tables=[]
  with pdfplumber.open(pdf_path) as pdf:
    for page_number, page in enumerate(pdf.pages,start=1):

      extracted=page.extract_tables()


      for table in extracted:
        # table은 2차원 배열이기 때문에 table[1:]는 내용, table[0]는 column명
        # if table=[[ ['Name', 'Age', 'City'],
                    # ['Alice', '30', 'New York'],
                    # ['Bob', '25', 'Los Angeles']]]
        df=pd.DataFrame(table[1:],columns=table[0])
        if df.empty:
          continue
        else:
          tables.append(df)

  for i,table in enumerate(tables):
    table.to_csv(f"/content/drive/MyDrive/사관학교챗봇/airforce_images/extracted_table_{i}.csv",index=False)
    print("Success")

extract_table(airforce_pdf)
```

```python
def extract_text(pdf_path):
  tables=[]
  with pdfplumber.open(pdf_path) as pdf:
    for page_number, page in enumerate(pdf.pages,start=1):

      extracted_table=page.extract_tables()
      text=page.extract_text()

      file=open(f'/content/drive/MyDrive/사관학교챗봇/airforce_images/airforce_txt_{page_number}','w')
      file.write(text)
      file.close()

```
